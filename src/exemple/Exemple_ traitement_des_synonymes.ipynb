```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Initialisation
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

mot = "ate"

# Stemming (Il va juste couper ce qu'il peut, ici ça ne marche pas sur les irréguliers)
print(f"Stemming de '{mot}': {stemmer.stem(mot)}")

# Lemmatisation (Il consulte le dictionnaire et sait que 'ate' vient de 'eat')
# Il faut préciser 'v' pour verbe
print(f"Lemmatisation de '{mot}': {lemmatizer.lemmatize(mot, pos='v')}")
```
