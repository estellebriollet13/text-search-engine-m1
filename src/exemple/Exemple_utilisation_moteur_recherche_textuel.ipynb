```python
import sys
import os

sys.path.append(os.path.abspath(".."))
from lib.MoteurRechercheTextuel import MoteurRechercheTextuel 

if __name__ == "__main__":
    mini_corpus = {
        0: "Le hachage hachage hachage est rapide, très rapide",
        1: "Le tri par fusion est stable",
        2: "Un index inversé est efficace",
        3: "Le hachage et l'index inversé sont une solution, mais le processus n'est pas toujours rapide",
        4: "Tri rapide vs tri stable",
        5: ""
    }

    moteur = MoteurRechercheTextuel()
    moteur.indexer_corpus(mini_corpus)

    print("--- 1. Corpus indexé ---")
    print(f"Nombre de documents : {len(moteur.documents)}")
    print(f"Taille du document 0 : {moteur.tailles_documents[0]} mots")
    print(f"Taille du document vide : {moteur.tailles_documents[5]} mot")

    print("\n--- 2. Vérification de l'index inversé ---")
    print(f"Posting list pour 'hachage' : {moteur.index.get('hachage')}")
    print(f"Posting list pour 'rapide'  : {moteur.index.get('rapide')}")
    print(f"Posting list pour 'tri'     : {moteur.index.get('tri')}")

    print("\n--- 3. Requêtes AND ---")
    requetes_and = [
        "hachage rapide",
        "tri stable",
        "index inversé",
        "mot absent",
        ""
    ]
    for requete in requetes_and:
        print(f"{requete!r:<18} -> {moteur.requete_and(requete)}")

    print("\n--- 4. Recherche TF-IDF ---")
    requete = "hachage rapide"
    resultats = moteur.chercher_tfidf(requete)
    print(f"Résultats pour {requete!r}, triés par score décroissant :")
    for doc_id, score in resultats:
        print(f" - Doc {doc_id} | score={score:.4f} | {moteur.documents[doc_id]}")

    print("\n--- 5. Recherche sans résultat ---")
    print(moteur.chercher_tfidf("hachage inconnu"))
```
